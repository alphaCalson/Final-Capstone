```python
import pandas as pd
import re
from googletrans import Translator
import time

# Function to clean the content (removing special characters)
def clean_content(text):
    return re.sub(r'[^A-Za-z\s]', '', text)



# Initialize the Google Translate API client
translator = Translator()
# Load the South African dataset
sa_dataset = pd.read_csv("/content/SA_XenoDataset_2017-2022.csv")



# Create 'id' column from the index
sa_dataset_transformed = pd.DataFrame()
sa_dataset_transformed['id'] = sa_dataset.index.astype(str)
sa_dataset_transformed['content'] = sa_dataset['CleanedTweet'].fillna('').astype(str)
sa_dataset_transformed['content'] = sa_dataset_transformed['content'].apply(clean_content)

# Load progress if a file already exists
try:
    completed = pd.read_csv('progress.csv')
    completed_ids = set(completed['id'])
except FileNotFoundError:
    completed = pd.DataFrame(columns=['id', 'content', 'lang'])
    completed_ids = set()

# Define a batch size
batch_size = 50

# Function to detect languages in batches and save progress
def detect_languages_batch(data, start=0, batch_size=50):
    batch_data = data[start:start + batch_size]
    for i, row in batch_data.iterrows():
        if row['id'] not in completed_ids:  # Skip if already processed
            try:
                # Detect language
                detected_lang = translator.detect(row['content']).lang
            except Exception:
                detected_lang = 'unknown'

            # Add detected language to the completed DataFrame
            completed.loc[len(completed)] = [row['id'], row['content'], detected_lang]

            # Save progress every batch
            if (i + 1) % batch_size == 0:
                completed.to_csv('progress.csv', index=False)
                print(f'Saved progress for batch ending at row {i+1}')

# Batch process the language detection
for start in range(0, len(sa_dataset_transformed), batch_size):
    detect_languages_batch(sa_dataset_transformed, start, batch_size)
    time.sleep(1)  # Avoid overwhelming the API
    print(f'Processed batch starting at row {start}')

# Save final dataset
completed.to_csv('transformed_sa_dataset_with_google_lang.csv', index=False)

# View the first few rows of the result
print(completed.head())
```
